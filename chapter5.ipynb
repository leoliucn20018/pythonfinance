```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
np.random.seed(1000)
y = np.random.standard_normal(20)
```

```python
x = np.arange(len(y))
```

```python
plt.plot(x,y)
```

```python
plt.plot(y)
```

```python
plt.plot(y.cumsum())
```

```python
plt.grid(True)
plt.plot(y.cumsum())
plt.axis('tight')
```

```python
plt.plot(y.cumsum())
plt.grid(True)
plt.xlim(-1,20)
plt.ylim(np.min(y.cumsum())-1,np.max(y.cumsum())+1)
```

```python
plt.figure(figsize=(7,4))
plt.plot(y.cumsum(),'b--',lw=1.5)
plt.plot(y.cumsum(),'ro')
plt.grid(True)
plt.axis('tight')
plt.xlabel('index')
plt.ylabel('value')
plt.title('A simple plot')
```

```python
np.random.seed(2000)
y = np.random.standard_normal((20,3)).cumsum(axis=0)
```

```python
plt.figure(figsize=(7,4))
plt.plot(y,lw=1.5)
plt.plot(y,'ro')
plt.grid(True)
plt.axis('tight')
plt.xlabel('index')
plt.ylabel('value')
plt.title('A Simple Plot')
```

```python
plt.figure(figsize = (7,4))
plt.plot(y[:,0],label = 'label0')
plt.plot(y[:,1],label='label1')
plt.plot(y[:,2],label='label2')
plt.plot(y,'ro')
plt.plot(y,'r--')
plt.grid(True)
plt.legend(loc=0)
plt.axis('tight')
plt.xlabel("index")
plt.ylabel("value")
plt.title("A simple plot")
```

```python
y[:,0] = y[:,0]*100
plt.figure(figsize = (7,4))
plt.plot(y[:,0],label = 'label0')
plt.plot(y[:,1],label='label1')
plt.plot(y[:,2],label='label2')
plt.plot(y,'ro')
plt.plot(y,'r--')
plt.grid(True)
plt.legend(loc=0)
plt.axis('tight')
plt.xlabel("index")
plt.ylabel("value")
plt.title("A simple plot")
```

```python
fig, ax1 = plt.subplots()
plt.plot(y[:,0],'b',lw = 1.5,label='1st')
plt.plot(y[:,0],'ro')
plt.grid(True)
plt.legend(loc=8)
plt.axis('tight')
plt.xlabel("index")
plt.ylabel("value 1st")
plt.title("a simple plot")
ax2 = ax1.twinx()
plt.plot(y[:,1],'g',lw=1.5,label='2nd')
plt.plot(y[:,1],'yo')
plt.legend(loc= 2)
plt.ylabel('value 2nd')
```

```python
plt.figure(figsize = (9,4))
plt.subplot(121)
plt.plot(y[:,0],'b',lw =0.5 ,label='1st')
plt.plot(y[:,0],'ro')
plt.subplot(122)
plt.bar(np.arange(len(y)),y[:,0],width =0.5 ,color='r',label='1st')
 
```

```python
y = np.random.standard_normal((1000,2))
```

```python
plt.figure(figsize = (7,5))
plt.plot(y[:,0],y[:,1],'ro')
plt.grid(True)
```

```python
plt.figure(figsize = (7,5))
plt.scatter(y[:,0],y[:,1],marker='o')
plt.grid(True)
```

```python
plt.figure(figsize = (7,5))
plt.scatter(y[:,0],y[:,1],c='r',marker='o')
plt.grid(True)
```

```python
plt.figure(figsize = (7,5))
plt.hist(y,label=['1st','2nd'],stacked = True,bins=20)
plt.grid(True)
plt.legend(loc=0)
```

```python
fig,ax = plt.subplots(figsize=(7,4))
plt.boxplot(y)
plt.grid(True)
plt.setp(ax,xticklabels=['1st','2nd'])
```
